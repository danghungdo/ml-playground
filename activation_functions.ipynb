```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
e = math.e
data = [1, 5, -4, 3, -2]
```

# Sigmoid Fuction
### $\operatorname{sigmoid}(x)=\frac{1}{1+e^{-x}}$

```python
def sigmoid(x):
    return 1 / (1 + e ** (-x))
```

```python
print([sigmoid(x) for x in data])
```

```python
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
def visualize(func1, func1_derivative):
    x = np.linspace(-5, 5, 100)
    y = func1(x)
    derivative = func1_derivative(x)
    plt.plot(x, y, label=func1.__name__, color='blue')
    plt.plot(x, derivative, label=func1_derivative.__name__, color='red')
    plt.axvline(x=0, c="black")
    plt.axhline(y=0, c="black")
    plt.xlabel('Input')
    plt.ylabel('Output')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
visualize(sigmoid, sigmoid_derivative)
```

# Tanh Function
### $\operatorname{tanh}(x)=\frac{2}{1+e^{-2x}}-1$

```python
def tanh(x):
    return (2 / (1 + e ** (-2 * x))) - 1
```

```python
print([tanh(x) for x in data])
```

```python
def tanh_derivative(x):
    return 1 - tanh(x) ** 2
```

```python
visualize(tanh, tanh_derivative)
```

# Softplus Function
### $\operatorname{Softplus}(x)=\log \left(1+e^x\right)$

```python
def softplus(x):
    return np.log(1 + e ** x)
```

```python
print([softplus(x) for x in data])
```

```python
def softplus_derivative(x):
    return 1 / (1 + e ** (-x))
```

```python
visualize(softplus, softplus_derivative)
```

# ReLU Function
### $\operatorname{relu}(x)= \begin{cases}0 & \text { if } x \leq 0 \\ x & \text { if } x>0\end{cases}$

```python
def relu(x):
    return np.maximum(x, 0)
```

```python
print([relu(x) for x in data])
```

```python
def relu_derivative(x):
    return np.where(x <= 0, 0, 1)
```

```python
visualize(relu, relu_derivative)
```

# LeakyReLU Function
### $\operatorname{LeakyRelu}(x)= \begin{cases}0.01 x & \text { if } x \leq 0 \\ x & \text { if } x>0\end{cases}$

```python
def leakyrelu(x):
    return np.maximum(x, 0.01 * x)
```

```python
print([leakyrelu(x) for x in data])
```

```python
def leakyrelu_derivative(x):
    return np.where(x > 0, 1, 0.01)
```

```python
visualize(leakyrelu, leakyrelu_derivative)
```

# ELU Function
### $E L U(x)=\left\{\begin{array}{cc}\alpha\left(e^x-1\right) & \text { if } x \leq 0 \\ x & \text { if } x>0\end{array}\right.$

```python
def elu(x, alpha = 1):
    return np.where(x > 0, x, alpha * (e ** x - 1))
```

```python
print(elu(np.array(data)))
```

```python
def elu_derivative(x, alpha = 1):
    return np.where(x > 0, 1, elu(x, alpha) + alpha)
```

```python
visualize(elu, elu_derivative)
```

# PReLU Function
### $ \operatorname{PReLU}(x)=\left\{\begin{array}{cc}\alpha x & \text { if } x \leq 0 \\ x & \text { if } x>0\end{array}\right.$

```python
def prelu(x, alpha = 0.1):
    return np.where(x > 0, x, x * alpha)
```

```python
print(prelu(np.array(data)))
```

```python
def prelu_derivative(x, alpha = 0.1):
    return np.where(x > 0, 1, alpha)
```

```python
visualize(prelu, prelu_derivative)
```

# Swisch Function
### $\operatorname{Swish}(x)=\frac{x}{1+e^{-\beta x}}=x \operatorname{sigmoid}(\beta x)$

```python
def swish(x, beta = 1):
    return x / (1 + e ** (-beta * x))
```

```python
print([swish(x) for x in data])
```

```python
def swish_derivative(x, beta = 1):
    return sigmoid(x) + x * sigmoid(x) * (1 - sigmoid(x))
```

```python
visualize(swish, swish_derivative)
```
